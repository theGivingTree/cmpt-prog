```python
n, m = int(input()), int(input())
bufr, bufc = 2*n+1, (4*n+4)*m;
buf = [[' ']*bufc for _ in range(bufr)]
```

```python
# init
x, y = bufr-1, -1
d = 0,1
f = 0
def r1(d): return -d[1],d[0]
def it(k):
    global d, x, y, f, e
    for _ in range(2*k if f else 4*k):
        x+=d[0];y+=d[1]
        print("%d %d" %(x,y))
        buf[x][y] = '|' if f else '-'
    buf[x][y] = '+'
    d = (r1(r1(r1(d))) if e else r1(d))
    f=1-f
```

```python
for _ in range(m):
    e=0
    for g in range(n, 0, -1):it(g)
    e=1
    it(1)
    for g in range(1, n+1, 1):it(g)
    it(n)
    d=r1(r1(d))
for i in range(bufr):
    print(''.join(buf[i]))
```
